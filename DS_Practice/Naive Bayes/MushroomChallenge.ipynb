```python
def prior_probab(y_train,label):
    #refer_notes
    total_examples=y_train.shape[0]
    class_examples=np.sum(y_train==label)
    
    return (class_examples/float(total_examples))
    
```

```python
def conditional_probab(X_train,y_train,feature_col,feature_val,label):
    x_filter=X_train[y_train==label]
    numerator=np.sum(x_filter[:,feature_col]==feature_val)
    denominator=np.sum(y_train==label)
    
    return numerator/float(denominator) 
```

```python
#find posterior probability
```

```python
def predict(x_train,y_train,x_test): #predicting class
    """Xtest is a single testing point"""
    classes=np.unique(y_train)
    n_features=x_train.shape[1]
    post_probs=[] #for all classes , each post probabs
    #compute posterior for each class
    for label in classes:
        #post_c=likelihood*prior
        likelihood=1.0
        for f in range(n_features):
            cond=conditional_probab(x_train,y_train,f,x_test[f],label)
            likelihood*=cond
        prior=prior_probab(y_train,label)
        post=likelihood*prior
        post_probs.append(post)
        pred=np.argmax(post_probs)
        
        return pred
```

```python
def score(X_train,y_trainy,x_test,y_test):
    pred=[]
    for i in range(x_test.shape[0]):
        pred_label=predict(X_train,y_train,x_test[i])
        pred.append(pred)
    pred=np.array(pred)
    
    accuracy=np.sum(pred==y_test)/y_test.shape[0]
    
    return accuracy*100
```
