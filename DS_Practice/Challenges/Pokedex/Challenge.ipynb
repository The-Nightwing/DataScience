```python
from pathlib import Path
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
from keras.applications.resnet50 import preprocess_input,ResNet50,decode_predictions
```

```python
dirs = p.glob("*")

image_data = []
labels = []
labels_dict = {"Squirtle":0,"Spearow":1,"Psyduck":2,"Pikachu":3,"Meowth":4,"Fearow":5,"Dratini":6,"Charmander":7,"Bulbasaur":8,"Aerodactyl":9}
label_to_pokemon = {0:"Squirtle",1:"Spearow",2:"Psyduck",3:"Pikachu",4:"Meowth",5:"Fearow",6:"Dratini",7:"Charmander",8:"Bulbasaur",9:"Aerodactyl"}
```

```python
for folder_name in dirs:
    pokemon_name = str(folder_name)
    
    for img in folder_name.glob("*.jpg"):
        img = image.load_img(img,target_size = (224,224))
        img_array = image.img_to_array(img)
        img = np.expand_dims(img,axis=0)
        img = preprocess_input(img)
        image_data.append(img_array)
        labels.append(labels_dict[pokemon_name])
```

```python
X = np.array(image_data)
y = np.array(labels)
```

```python
X = X/255.0
```
