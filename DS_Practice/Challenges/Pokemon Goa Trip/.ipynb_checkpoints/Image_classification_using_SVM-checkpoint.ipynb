```python
import numpy as np
import os 
import cv2
import pandas as pd
from pathlib import Path
from keras.preprocessing import image
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
train = pd.read_csv("Train/train.csv").values
# test = pd.read_csv("Test (5)/test.csv").values 
```

```python
ImageId = train[:,0]
train_labels = train[:,1]
train_labels[train_labels == 'Pikachu'] = 1
train_labels[train_labels == 'Charmander'] = -1
train_labels[train_labels == 'Bulbasaur'] = -1
```

```python
print(train_labels)
```

```python
images_data = []
for img in ImageId:
    image = cv2.imread("Train/Images/"+img)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image  = cv2.resize(image,(32,32))
    image = image.reshape((-1,))
    images_data.append(image)
```

```python
images_data = np.array(images_data)
print(images_data.shape)
```

```python
train_labels = np.array(train_labels)
```

```python
class SVM:
   
    def __init__(self,C=1.0):
        self.C = C
        self.W = 0
        self.b = 0
        
    def hingeLoss(self,W,b,X,Y):
        loss  = 0.0
        
        loss += .5*np.dot(W,W.T)
        
        m = X.shape[0]
        #print(m)
        for i in range(m):
            ti = Y[i]*(np.dot(W,X[i].T)+b)
            #print(Y[i])
            #print(W)
            #print(X[i].T)
            #print(ti)
            loss += self.C *max(0,(1-ti))
            #print(loss)
            
        return loss[0][0]
    
    def fit(self,X,Y,batch_size=100,learning_rate=0.00001,maxItr=300):
        
        no_of_features = X.shape[1]
        no_of_samples = X.shape[0]
        
        n = learning_rate
        c = self.C
        
        #Init the model parameters
        W = np.zeros((1,no_of_features))
        bias = 0
        
        #Initial Loss
        
        #Training from here...
        # Weight and Bias update rule that we discussed!
        losses = []
        
        for i in tqdm(range(maxItr)):
            #Training Loop
            
            l = self.hingeLoss(W,bias,X,Y)
            losses.append(l)
            ids = np.arange(no_of_samples)
            np.random.shuffle(ids)
            
            #Batch Gradient Descent(Paper) with random shuffling
            for batch_start in range(0,no_of_samples,batch_size):
                #Assume 0 gradient for the batch
                gradw = 0
                gradb = 0
                
                #Iterate over all examples in the mini batch
                for j in range(batch_start,batch_start+batch_size):
                    if j<no_of_samples:
                        i = ids[j]
                        ti =  Y[i]*(np.dot(W,X[i].T)+bias)
                        
                        if ti>1:
                            gradw += 0
                            gradb += 0
                        else:
                            gradw += c*Y[i]*X[i]
                            gradb += c*Y[i]
                            
                #Gradient for the batch is ready! Update W,B
                W = W - n*W + n*gradw
                bias = bias + n*gradb
                
        
        self.W = W
        self.b = bias
        return W,bias,losses      
```

# Without normalization

```python

mySVM = SVM()
w,b,losses = mySVM.fit(images_data,train_labels,maxItr=1000)
```

```python
plt.figure(figsize=(12,12))
plt.plot(losses)
```

# With normalization

```python
mySVM = SVM()
w,b,losses = mySVM.fit(images_data/255.0,train_labels,maxItr=1000)
```

```python
plt.figure(figsize=(12,12))
plt.plot(losses)
```
