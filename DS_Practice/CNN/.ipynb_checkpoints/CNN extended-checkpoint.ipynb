```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img
```

```python
img=np.array(load_img("./pic.jpeg",target_size=(200,200)))
```

```python
img.shape
```

```python
gray=img.mean(axis=2)
```

```python
plt.imshow(gray,cmap="gray")
```

```python
plt.imshow(img)
```

```python
fil=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
fil
```

```python
padded_img=np.pad(gray,1) #padding of 1
```

```python
padded_img.shape
```

```python
result=np.empty_like(gray)  #input se farq ni pdta, empty like ka mtlb 
```

```python
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut=padded_img[row:row+3,col:col+3] #3-->size of filter
        result[row,col]=np.abs(np.mean(fil*cut)) #to take rising and falling as same
        
```

```python
plt.imshow(result,cmap="gray")
```
