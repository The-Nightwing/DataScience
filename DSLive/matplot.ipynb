```python
theta=np.linspace(-np.pi,np.pi,100)
```

```python
x=np.cos(theta)
```

```python
y=np.sin(theta)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
x=np.linspace(-40*np.pi,40*np.pi,1000) #20 revolutions as 2pi covers one revolution
l=x
t=l*np.sin(x)
r=l*np.cos(x)
plt.figure(figsize=(6,6))
plt.plot(t,r,marker="*")
```
