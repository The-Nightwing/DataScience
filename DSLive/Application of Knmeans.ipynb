```python
X=img_array.reshape(-1,3)
```

```python
X.shape
```

```python
#ab isme se 6 color nikalenge , as in 6 cluster centers honge jinke corresponding 6 colors ka cluster hoga
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
model.fit(X)
```

```python
pallets=model.cluster_centers_.astype(np.uint8) #6 color ka color pallets
```

```python
pallets
```

```python
#because the collor pallets will be in 
#perfect integer and not decimals
```

```python
plt.imshow(pallets.reshape(1,6,3)) 
#1 row 6 column of colors and 3 ki depth(rgb wali)
```

```python
model.labels_[:100]
```

```python
output=pallets[model.labels_]# har label ka color dedega
```
