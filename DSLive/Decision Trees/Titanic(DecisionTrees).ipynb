```python
def entropy(col):
    
    data_=np.unique(col,return_counts=True)
    n=col.shape[0]
    entro=0.0
    for i in data_[1]: #data_[1] has frequencies of data in that column
        p=i/n #probability
        entro+=-1.0*p*np.log2(p)
        
    return entro
```

```python
def divide_data(x_data,fkey,fval): #fvalue is the threshhold value for deciding to throw data inn left or right, fkey is column name.
    
    x_right=pd.DataFrame([],columns=x_data.columns)
    x_left=pd.DataFrame([], columns=x_data.columns)
    
    for ix in range(x_data.shape[0]):
        
        val=x_data[fkey].loc[ix] #to take that particluar value of fkey column and ix row , issi value se fval check krenge

        if val<fval:
            x_left=x_left.append(x_data.loc[ix]) #append whole ix row  
        else: 
            x_right=x_right.append(x_data.loc[ix])
        
    return x_left,x_right
```

```python
def info_gain(x_data,fkey,fval):
    
    left,right=divide_data(x_data,fkey,fval)
    #now find out percentage of data in both left and right
    l=float(left.shape[0])/x_data.shape[0]
    r=float(right.shape[0])/x_data.shape[0]
    
    #agar ek hi side p sare examples chle gye mtlb min hogya info_gain aur return krdo
    
    if left.shape[0]==0 or right.shape[0]==0:
        return -99999999
    
    i_gain=entropy(x_data)-(l*entropy(left.Survived)+r*entropy(right.Survived))
    
    return i_gain
```

```python
class DecisionTree:
    
    def __init__(self,depth=0,max_depth=5):
        self.left=None
        self.right=None
        self.fkey=None #about which feature im splitting
        self.fval=None #threshhold for left right distribution
        self.max_depth=max_depth
        self.depth=depth
        
    def train(self,X_train):
        features=['Pclass','Sex','Age','SibSp','Parch']
        info_gains=[] #to store all info gains of the goven columns
        
        for ix in features:
            i_gain=info_gain(X_train,ix,X_train[ix].mean())
            info_gains.append(i_gain)
            
        self.fkey=features[np.argmax(info_gains)]
        self.fval=X_train[self.fkey].mean()
        
        #split
        x_left,x_right=divide_data(X_train,self.fkey,self.fval)
        x_left=x_left.reset_index(drop=True) #this is because when we divide, if lets say from 1,2,3,4,5 rows--> 2,3,4 goes in left so the numbering should be as 1-2-3 and not as 2,3,4 
        x_right=x_right.reset_index(drop=True)
        
        print('making tree, feature:'+self.fkey)
        
        #when leaf node is encountered
        if x_left.shape[0]==0 or x_right.shape[0]==0: #terminal node
            if X_train.Survived.mean()>=0.5: #zyada +ve cases than negative
                self.target="Survive"
            else:
                self.target="Dead"
            return
                
        #if depth>=max_Depth
        if self.depth >= self.max_depth:   
            if X_train.Survived.mean()>=0.5:
                self.target="Survive"
                
            else:
                self.target="Dead"
            
            return
                
        #recursive case to build the tree
        self.left=DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.left.train(x_left)
        
        self.right=DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.right.train(x_right)
        
        #checking at every node too
        if X_train.Survived.mean()>=0.5:#zyada +ve cases than negative
            self.target="Survive"
        else:
            self.target="Dead"   
        return
    
    def predict(self,test):
        
        if test[self.fkey] > self.fval: #matlab right m jaao
            if self.right is None: #leaf node or we can say if the node at which we are has no child.
                return self.target
            return self.right.predict(test) #keep exploring right subtree
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
        
        
        
        
```
